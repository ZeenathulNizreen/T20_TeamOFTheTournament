```python
import pandas as pd
import json
```

### Match Results ###

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data=json.load(f)
  
```

```python
df_match= pd.DataFrame(data[0]['matchSummary'])
df_match.tail()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard' : 'match_id'}, axis=1, inplace=True)
```

```python
df_match.head()
```

```python
match_ids_dict = {}

for index,row in df_match.iterrows():
    key1 = row['team1'] + " Vs "+row['team2']
    key2 = row['team2'] + " Vs "+row['team1']
    
    match_ids_dict[key1]=row['match_id']
    match_ids_dict[key2]=row['match_id']
```

```python
df_match.head()
```

```python
df_match.to_csv('t20_csv_files/dim_match_summary.csv')
```

### Batting Summary ###


```python
with open('t20_json_files/t20_wc_batting_summary.json') as b:
    batting_data=json.load(b)
```

```python
df_batting= pd.DataFrame(batting_data[0]['battingSummary'])
df_batting
```

```python
df_batting= pd.DataFrame(batting_data[1]['battingSummary'])
df_batting
```

```python
all_records=[]

for rec in batting_data:
    all_records.extend(rec['battingSummary'])
                           
df_batting =pd.DataFrame(all_records)
df_batting.tail()                          
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda a : "out" if len(a)>0 else "not_out")
df_batting
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(10)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda a: a.replace('â€',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda a: a.replace('\xa0',''))
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
```

```python
df_batting
```

```python
df_batting.shape
```

```python
df_batting.to_csv('t20_csv_files/fact_batting_summary.csv', index=False)
```

### Player Info
